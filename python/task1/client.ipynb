```python
from abc import ABC,abstractmethod
import os
import json
import csv

class BaseEntity(ABC):
    @abstractmethod
    def to_json(self):
        pass
    def to_csv(self):
        pass
            
class Street(BaseEntity):
    def __init__(self,city,district):
        self.city=city
        self.district=district
                
    def to_json(self):
        os.makedirs("Street",exist_ok=True)
        if not os.path.exists("Street/data.json"):
            with open("Street/data.json","w") as f:
                json.dump([{"city":self.city, "district":self.district}],f,indent=4)
        else:
            
            with open("Street/data.json","r") as f:
                exist_data=json.load(f)
            exist_data.append([{"city":self.city, "district":self.district}])
            
            with open("Street/data.json","w") as f:
                json.dump(exist_data,f,indent=4)        
                
                                        
    def to_csv(self):
        os.makedirs("Street",exist_ok=True)
        if not os.path.exists("Street/data.csv"):
            with open("Street/data.csv","w",newline="") as f:
                writer=csv.writer(f)
                writer.writerow(["city","district"])
                writer.writerow([self.city,self.district])
        else:
            with open("Street/data.csv", "r") as f:
                reader = csv.reader(f)
                existing_data = list(reader)

            # Add the new data to the existing data
            existing_data.append([self.city, self.district])

            # Write the updated data back to the CSV file
            with open("Street/data.csv", "w", newline="") as f:
                writer = csv.writer(f)
                writer.writerows(existing_data) 
                
                
                
class Student(BaseEntity):
    def __init__(self,first_name,last_name):
        self.first_name=first_name
        self.last_name=last_name
                
    def to_json(self):
        os.makedirs("Student",exist_ok=True)
        if not os.path.exists("Student/data.json"):
            with open("Student/data.json","w") as f:
                json.dump([{"first_name":self.first_name, "last_name":self.last_name}],f,indent=4)
        else:
            
            with open("Student/data.json","r") as f:
                exist_data=json.load(f)
            exist_data.append([{"first_name":self.first_name, "last_name":self.last_name}])
            
            with open("Student/data.json","w") as f:
                json.dump(exist_data,f,indent=4)        
                
                                        
    def to_csv(self):
        os.makedirs("Student",exist_ok=True)
        if not os.path.exists("Student/data.csv"):
            with open("Student/data.csv","w",newline="") as f:
                writer=csv.writer(f)
                writer.writerow(["first_name","last_name"])
                writer.writerow([self.first_name,self.last_name])
        else:
            with open("Student/data.csv", "r") as f:
                reader = csv.reader(f)
                existing_data = list(reader)

            # Add the new data to the existing data
            existing_data.append([self.first_name, self.last_name])

            # Write the updated data back to the CSV file
            with open("Student/data.csv", "w", newline="") as f:
                writer = csv.writer(f)
                writer.writerows(existing_data) 

                                 
        
        
```

```python
import random
import string

def randomstr(num):    
    characters=string.ascii_letters +string.digits
    return ''.join(random.choice(characters) for _ in range(num)) 
    
```

```python
for i in range(100):
    num=4
    st=Street(randomstr(num),randomstr(num))
    st.to_json()
    st.to_csv()
    
    stu=Student(randomstr(num),randomstr(num))
    stu.to_json()
    stu.to_csv()
```
