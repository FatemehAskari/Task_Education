```python
class SearchFile :
    def __init__(self,root_dir):
        self.root_dir=root_dir
        
    def FindnameFile(self,file_name):
        result=[]
        for root,dirs,files in os.walk(self.root_dir):
            for file in files:
                if file.startswith(file_name):
                    result.append(os.path.join(root,file))
        return result
    
    def FindTypeFile(self,typefile):
        result=[]
        for root,dirs,files in os.walk(self.root_dir):
            for file in files:
                if len(os.path.basename(file).split('.'))>1 and os.path.basename(file).split('.')[1]==typefile:                    
                    result.append(os.path.join(root,file))
        return result
    
    def FindContent(self,content):
        result=[]
        for root,dirs,files in os.walk(self.root_dir):
            for file in files:
                if len(os.path.basename(file).split('.'))==2 and os.path.basename(file).split('.')[1]=='txt':
                    path=os.path.join(root,file)
                    with open(str(os.path.join(root,file)),"r",encoding="utf-8") as file:
                        file_contenet=file.read()
                    if content in str(file_contenet):
                        result.append(path)
        return result                    
                                               
                    
```

```python
import os
import datetime


def find_type(file):
    _, file_extension = os.path.splitext(file)
    file_type = file_extension[1:]
    return file_type 

def find_name(file):
    base_name = os.path.splitext(os.path.basename(file))[0]
    return base_name 
def find_content(file,file_type,root):
    accept_type=['txt','csv']
    file_content=""
    if file_type in accept_type:
        path1=os.path.join(root,file)
        try:            
            with open(str(path1), "r", encoding="utf-8") as file:
                file_content = file.read()
        except UnicodeDecodeError:
            print(f"Error decoding file: {file}")
            file_content = ""  
    return file_content

def find_time(file, root):
    path1 = os.path.join(root, file)
    creation_time = os.path.getctime(path1)
    creation_datetime = datetime.datetime.fromtimestamp(creation_time)
    return str(creation_datetime)                   
```

```python
import json

def tile_json():
    data = {
        "MaxId": "My Data",
        "node": []
    }

    # Write the data to a JSON file
    with open("data.json", "w") as file:
        json.dump(data, file, indent=2)

def create_json(root_dir):
    id1=0
    tile_json()
    with open('data.json','r') as file:
        data=json.load(file)                    
    for root, dirs, files in os.walk(root_dir):
        name_root=find_name(root)
        type_root='folder'        
        content_root=""
        id_root=id1
        date_root=find_time(root,"")
        child_root=[]
        for file in files:
            id1+=1
            child_root.append(id1)
                                    
            name_child=find_name(file)
            type_child=find_type(file)
            content_child=find_content(file,type_child,root)
            id_child=id1
            date_child=find_time(file,root)
            # date_child=""
            child_child=[]
            
            data['node'].append({
                "id":id_child,
                "name":name_child,
                "time":date_child,
                "type":type_child,
                "content":content_child,
                "child":child_child
            })
        data['node'].append({
        "id":id_root,
        "name":name_root,
        "time":date_root,
        "type":type_root,
        "content":content_root,
        "child":child_root
        })
    data['MaxId']=id1    
    with open("data.json", "w") as file:
        json.dump(data, file, indent=2)    
```
